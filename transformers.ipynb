```python
from transformers import pipeline

# Load a pre-trained BERT model for sentiment analysis
classifier = pipeline("text-classification", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Example text input
texts = [
    "I love this product! It's amazing.",
    "The service was terrible and disappointing."
]

# Perform classification
results = classifier(texts)

# Print results
for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Prediction: {result['label']}, Confidence: {result['score']:.4f}\n")
```

```python
import torch

# Fake word embeddings (each word has a vector of 4 numbers)
sentence = torch.tensor([
    [0.1, 0.2, 0.3, 0.4],  # "I"
    [0.2, 0.1, 0.4, 0.3],  # "love"
    [0.3, 0.4, 0.1, 0.2]   # "pizza"
])

# Add batch dimension (1 sentence, 3 words, 4 features per word)
sentence = sentence.unsqueeze(0)

print(sentence.shape)  # Output: torch.Size([1, 3, 4])
```

```python
import torch.nn as nn

# Define weight matrices for Query, Key, and Value
embed_size = 4  # Embedding size of words
W_q = nn.Linear(embed_size, embed_size, bias=False)  # Query weight matrix
W_k = nn.Linear(embed_size, embed_size, bias=False)  # Key weight matrix
W_v = nn.Linear(embed_size, embed_size, bias=False)  # Value weight matrix

# Apply transformations
Q = W_q(sentence)  # Queries
K = W_k(sentence)  # Keys
V = W_v(sentence)  # Values

print(Q.shape, K.shape, V.shape)  # Output: torch.Size([1, 3, 4])
```

```python
# Compute attention scores (Q @ K^T)
scores = torch.matmul(Q, K.transpose(-2, -1))

print(scores.shape)  # Output: torch.Size([1, 3, 3])
print(scores)  # Shows the attention scores
```

```python
import torch.nn.functional as F

# Apply softmax to get attention weights
attention_weights = F.softmax(scores, dim=-1)

print(attention_weights.shape)  # Output: torch.Size([1, 3, 3])
print(attention_weights)  # Shows normalized attention values
```

```python
# Multiply attention weights with Values (V)
output = torch.matmul(attention_weights, V)

print(output.shape)  # Output: torch.Size([1, 3, 4])
print(output)  # Shows the transformed sentence
```

```python
import torch.nn as nn  # Neural network module

class SimpleTransformerBlock(nn.Module):
    def __init__(self, embed_size, heads):
        super(SimpleTransformerBlock, self).__init__()
        self.attention = nn.MultiheadAttention(embed_dim=embed_size, num_heads=heads)
        self.norm = nn.LayerNorm(embed_size)

    def forward(self, x):
        attn_output, _ = self.attention(x, x, x)  # Self-Attention
        x = self.norm(attn_output + x)  # Skip Connection + Normalization
        return x
```

```python
# Define model
embed_size = 4  # Same as word embedding size
heads = 2       # Number of attention heads

model = SimpleTransformerBlock(embed_size, heads)

# Run model on sentence
output = model(sentence)

print(output.shape)  # Output: torch.Size([1, 3, 4])
print(output)  # View transformed sentence
```
